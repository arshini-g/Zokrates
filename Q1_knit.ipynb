```python
def get_transaction_history(wallet_address, api_key):
    # Specify the Etherscan API endpoint for retrieving transaction history
    url = f"https://api.etherscan.io/api?module=account&action=txlist&address={wallet_address}&apikey={api_key}"
    try:
        response = requests.get(url)
        data = response.json()
        bal = 0
        remaining = 0
        if data['status'] == '1':
            transactions = data['result']
            
            for transaction in transactions:
                if transaction['to'] == wallet_address.lower():
                    remaining = int(transaction['value']) - int(transaction['gasPrice']) * int(transaction['gasUsed'])
                    #print("remaining balance1:", transaction['value'] - (transaction['gasPrice'] * transaction['gasUsed']))
                else:
                    remaining = -int(transaction['value']) - int(transaction['gasPrice']) * int(transaction['gasUsed'])
                    #print("remaining balance:", -int(transaction['value']) - int(transaction['gasPrice']) * int(transaction['gasUsed'])) 
                bal = bal + remaining
    
        else:
            print("API call failed. Error message:", data['message'])
        if bal < 0:
            print("Wallet Address:", wallet_address)
            print("Loss:", bal, "ETH")
        else:
            print("Wallet Address:", wallet_address)
            print("Profit:", bal, "ETH")
        print("----------------------------------------------------------")    

    except requests.exceptions.RequestException as e:
        print("Error occurred:", str(e))

```
